```python
from model import RUN_CSP
from csp_utils import CSP_Instance, Constraint_Language
from train import train

import data_utils
import tensorflow as tf

import random
from tqdm import tqdm
import time
import shutil
```

```python
def train_with_parameters(data_path, n_colors, model_dir, batch_size, epochs, t_max, limit=-1, delete_previous_model=True):
    if delete_previous_model:
        shutil.rmtree(model_dir, ignore_errors=True)

    language = Constraint_Language.get_coloring_language(n_colors)

    print('loading graphs...')
    names, graphs = data_utils.load_graphs(data_path, limit=limit)
    random.shuffle(graphs)
    print('Converting graphs to CSP Instances')
    instances = [CSP_Instance.graph_to_csp_instance(g, language, 'NEQ') for g in tqdm(graphs)]
    batches = CSP_Instance.batch_instances(instances, batch_size)

    # construct and train new network
    tf.keras.backend.clear_session()
    network = RUN_CSP(model_dir, language)

    time_start = time.perf_counter()
    train(network, batches, epochs=epochs, t_max=t_max)
    time_end = time.perf_counter()

    print(f"Total training time: {time_end - time_start:.2f} seconds")
```

```python
BATCH_SIZES = [1, 5, 10]
ITERATIONS = [10, 50, 100]
N_COLORS = 8
```

```python
for batch_size in BATCH_SIZES:
    for t_max in ITERATIONS:
        print(f"\n\n===== Training with batch size {batch_size} and t_max {t_max} =====")

        model_dir = f"models/RegAlloc/chess_GR64_t{t_max}_b{batch_size}_8colors"
        train_with_parameters(
            data_path="../../../chess/build/interference_graphs/GR64",
            n_colors=N_COLORS,
            model_dir=model_dir,
            batch_size=batch_size,
            epochs=5,
            t_max=t_max,
        )
```

```python
import evaluate

def validate_model(graphs, model_dir, attempts=100, t_max=500):
    tf.keras.backend.clear_session()
    network = RUN_CSP.load(model_dir)

    print('Converting graphs to CSP Instances')
    instances = [CSP_Instance.graph_to_csp_instance(g, network.language, 'NEQ') for g in tqdm(graphs)]

    evaluate.evaluate_boosted(
        network, instances, t_max, attempts=attempts
    )
```

```python
from greedy import greedyColoring

def validate_greedy(graphs, max_colors=8):
    print('Validating greedy coloring')
    success_count = 0
    for graph in tqdm(graphs):
        n_colors = greedyColoring(graph)
        success_count += (n_colors <= max_colors)

    print(f"Greedy coloring solved {success_count} out of {len(graphs)} problems")
```

```python
def validate_file(filename, model_dir, attempts=100, t_max=500):
    tf.keras.backend.clear_session()
    network = RUN_CSP.load(model_dir)

    graph = data_utils.load_dimacs_graph(filename)
    instance = CSP_Instance.graph_to_csp_instance(graph, network.language, 'NEQ')
    output_dict = network.predict_boosted(instance, iterations=t_max, attempts=attempts)

    n_conflicts = instance.count_conflicts(output_dict['assignment'])

    print(f"Conflicts: {n_conflicts}")
```
